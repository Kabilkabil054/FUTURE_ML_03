```python
import pandas as pd
```

```python
data = pd.read_csv("resumes.csv")
data.head()
```

```python
data['clean_resume'] = data['Resume_Text']
```

```python
data['clean_resume'] = data['clean_resume'].fillna("").astype(str)
```

```python
data['clean_resume'] = data['clean_resume'].str.lower()
```

```python
data['clean_resume'] = data['clean_resume'].str.replace('[^a-z ]', '', regex=True)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    cleaned = []
    for word in words:
        if word not in stop_words:
            cleaned.append(word)
    return " ".join(cleaned)

data['clean_resume'] = data['clean_resume'].apply(remove_stopwords)
```

```python
data[['Resume_Text', 'clean_resume']].head()
```

```python
job_role = "Data Scientist"
```

```python
job_skills = [
    "python",
    "machine learning",
    "pandas",
    "numpy",
    "sql"
]
```

```python
def extract_skills(text):
    matched = []
    for skill in job_skills:
        if skill in text:
            matched.append(skill)
    return matched
```

```python
data['matched_skills'] = data['clean_resume'].apply(extract_skills)
```

```python
data['score'] = data['matched_skills'].apply(len)
```

```python
def find_missing_skills(matched):
    return list(set(job_skills) - set(matched))

data['missing_skills'] = data['matched_skills'].apply(find_missing_skills)
```

```python
ranked_candidates = data.sort_values(by='score', ascending=False)
```

```python
ranked_candidates[['Resume_ID', 'matched_skills', 'missing_skills', 'score']].head(10)
```

```python
## Result
Resumes are screened by matching required job skills.
Each resume is scored based on matched skills.
Candidates are ranked, and missing skills are identified.
```
